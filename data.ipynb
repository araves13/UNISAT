```python
data = dict()
for file in os.listdir('data/navi'):
    if file.endswith(".csv"):
        data[file.replace(".csv","")] = pd.read_csv(os.path.join('data/navi', file),header=[1], sep=';', comment='#')

sdata = dict()
for file in os.listdir('data'):
    if file.endswith(".csv"):
        sdata[file.replace(".csv","")] = pd.read_csv(os.path.join('data', file),header=[1], sep=';', comment='#')
```

```python
rrrdf=rrrdf.sort_values(by='Date GPS')
```

```python
rrrdf['Time_Diff']=rrrdf['Date GPS'].diff().apply(lambda x: x/np.timedelta64(1, 's')).fillna(0).astype('int64')
```

```python

rrrdf['Altitude Diff']=rrrdf['Altitude, m'].diff().apply(lambda x: x).fillna(0).astype('float')

```

```python
rrrdf['Vert_Velocity, m/s']=rrrdf['Altitude Diff']/rrrdf['Time_Diff']
```
